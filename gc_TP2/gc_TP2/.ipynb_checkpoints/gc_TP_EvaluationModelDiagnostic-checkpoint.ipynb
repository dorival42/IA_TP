### 3.1 Vrais positifs, faux positifs, vrais négatifs et faux négatifs

Les statistiques les plus élémentaires à calculer à partir des prédictions du modèle sont les vrais positifs, les vrais négatifs, les faux positifs et les faux négatifs. 

Comme leur nom l'indique
- Vrai positif (TP) : Le modèle classe l'exemple comme positif et l'étiquette réelle est également positive.
- faux positif (FP) : le modèle classe l'exemple comme positif, **mais** l'étiquette réelle est négative.
- vrai négatif (TN) : Le modèle classe l'exemple comme négatif et l'étiquette réelle est également négative.
- faux négatif (FN) : Le modèle classe l'exemple comme négatif, **mais** l'étiquette est en fait positive.

Nous allons compter le nombre de TP, FP, TN et FN dans les données données données.  Toutes nos mesures peuvent être construites à partir de ces quatre statistiques.

### 3.3 Prévalence
Un autre concept important est la **prévalence**. 
* Dans un contexte médical, la prévalence est la proportion de personnes dans la population qui sont atteintes de la maladie (ou de l'affection, etc.). 
* En termes d'apprentissage automatique, il s'agit de la proportion d'exemples positifs. L'expression de la prévalence est la suivante :

$$prevalence = \frac{1}{N} \sum_{i} y_i$$

où $y_i = 1$ lorsque l'exemple est 'positif' (a la maladie).

### 3.5 PPV and NPV

En revanche, sur le plan diagnostique, la sensibilité et la spécificité ne sont pas utiles. La sensibilité, par exemple, nous indique la probabilité que notre test soit positif si la personne est déjà atteinte de la maladie. Ici, nous conditionnons la chose que nous aimerions découvrir (si le patient est atteint de la maladie) !

Il serait plus utile de connaître la probabilité que la personne soit atteinte de la maladie si notre test est positif. Cela nous amène à la valeur prédictive positive (VPP) et à la valeur prédictive négative (VPN).

- La valeur prédictive positive (VPP) est la probabilité que les sujets dont le test de dépistage est positif soient réellement atteints de la maladie.
- La valeur prédictive négative (VPN) est la probabilité que les sujets dont le test de dépistage est négatif n'aient vraiment pas la maladie.

Là encore, nous pouvons formuler ces valeurs en termes de vrais positifs, de vrais négatifs, de faux positifs et de faux négatifs : 

$$PPV = \frac{\text{vrais positifs}}{\text{vrais positifs} + \text{faux positifs}}$$ 

$$NPV = \frac{\text{vrais négatifs}}{\text{vrais négatifs} + \text{faux negatifs}}$$

### 3.6 Courbe ROC

Jusqu'à présent, nous sommes partis de l'hypothèse que les prévisions de notre modèle de `0.5` et plus devaient être considérées comme positives et que, dans le cas contraire, elles devaient être considérées comme négatives. Il s'agit toutefois d'un choix plutôt arbitraire. Pour s'en rendre compte, il suffit d'examiner une visualisation très instructive appelée courbe ROC (receiver operating characteristic).

La courbe ROC est créée en traçant le taux de vrais positifs (TPR) par rapport au taux de faux positifs (FPR) à différents seuils. Le point idéal se situe en haut à gauche, avec un taux de vrais positifs de 1 et un taux de faux positifs de 0. Les différents points de la courbe sont générés en modifiant progressivement le seuil.

Examinons cette courbe pour notre modèle :

````python
util.get_curve(y, pred, class_labels)
````

L'aire sous la courbe ROC est également appelée AUCROC ou C-statistique et constitue une mesure de la qualité de l'ajustement. Dans la littérature médicale, ce chiffre indique également la probabilité qu'un patient sélectionné au hasard et ayant présenté une pathologie ait un score de risque plus élevé qu'un patient n'ayant pas présenté l'événement. Cela résume les résultats du modèle pour tous les seuils et donne une bonne idée du pouvoir discriminant d'un modèle donné.

Utilisons la fonction métrique `sklearn` de `roc_auc_score` pour ajouter ce score à notre table de métriques.

## 4. Intervalles de Confiance

Bien sûr, notre ensemble de données n'est qu'un échantillon du monde réel, et les valeurs calculées pour toutes les mesures ci-dessus sont une estimation des valeurs réelles. Il serait bon de quantifier cette incertitude due à l'échantillonnage de notre ensemble de données. Pour ce faire, nous utiliserons des intervalles de confiance. Un intervalle de confiance à 95 % pour une estimation $\hat{s}$ d'un paramètre $s$ est un intervalle $I = (a, b)$ tel que 95 % du temps lorsque l'expérience est réalisée, la vraie valeur $s$ est contenue dans $I$. Plus concrètement, si nous réalisons l'expérience un grand nombre de fois, la fraction des expériences pour lesquelles $I$ contient le vrai paramètre tendra vers 95 %.

Si certaines estimations sont accompagnées de méthodes permettant de calculer analytiquement l'intervalle de confiance, des statistiques plus complexes, telles que AUC par exemple, sont difficiles à obtenir. Pour celles-ci, nous pouvons utiliser une méthode appelée *bootstrap*. Le bootstrap estime l'incertitude en rééchantillonnant l'ensemble des données avec remplacement. Pour chaque rééchantillonnage $i$, nous obtenons une nouvelle estimation, $\hat{s}_i$. Nous pouvons alors estimer la distribution de $\hat{s}$ en utilisant la distribution de $\hat{s}_i$ pour nos échantillons bootstrap.

Dans le code ci-dessous, nous créons des échantillons bootstrap et calculons les AUCs des échantillons à partir de ces échantillons. Notez que nous utilisons un échantillonnage aléatoire stratifié (échantillonnage des classes positives et négatives séparément) pour nous assurer que les membres de chaque classe sont représentés. 

````python
def bootstrap_auc(y, pred, classes, bootstraps = 100, fold_size = 1000):
    statistics = np.zeros((len(classes), bootstraps))

    for c in range(len(classes)):
        df = pd.DataFrame(columns=['y', 'pred'])
        df.loc[:, 'y'] = y[:, c]
        df.loc[:, 'pred'] = pred[:, c]
        # get positive examples for stratified sampling
        df_pos = df[df.y == 1]
        df_neg = df[df.y == 0]
        prevalence = len(df_pos) / len(df)
        for i in range(bootstraps):
            # stratified sampling of positive and negative examples
            pos_sample = df_pos.sample(n = int(fold_size * prevalence), replace=True)
            neg_sample = df_neg.sample(n = int(fold_size * (1-prevalence)), replace=True)

            y_sample = np.concatenate([pos_sample.y.values, neg_sample.y.values])
            pred_sample = np.concatenate([pos_sample.pred.values, neg_sample.pred.values])
            score = roc_auc_score(y_sample, pred_sample)
            statistics[c][i] = score
    return statistics

statistics = bootstrap_auc(y, pred, class_labels)
````

Nous pouvons maintenant calculer les intervalles de confiance à partir des statistiques d'échantillon que nous avons calculées.

````python
util.print_confidence_intervals(class_labels, statistics)
````

Comme vous pouvez le constater, nos intervalles de confiance sont beaucoup plus larges pour certaines classes que pour d'autres. La hernie, par exemple, a un intervalle autour de (0,30 - 0,98), ce qui indique que nous ne pouvons pas être certains qu'elle est meilleure que le hasard (à 0,5).

## 5. Courbe Précision-Rappel


La précision-rappel est une mesure utile du succès de la prédiction lorsque les classes sont très déséquilibrées. 

Dans la recherche d'informations
- La précision est une mesure de la pertinence des résultats, qui équivaut à la VPP définie précédemment. 
- Le rappel est une mesure du nombre de résultats réellement pertinents renvoyés, ce qui équivaut à la mesure de sensibilité définie précédemment.

La courbe précision-rappel (PRC) montre le compromis entre la précision et le rappel pour différents seuils. Une zone élevée sous la courbe représente à la fois un rappel et une précision élevés, où une précision élevée est liée à un faible taux de faux positifs et un rappel élevé à un faible taux de faux négatifs. 

Des scores élevés pour les deux indiquent que le classificateur renvoie des résultats exacts (précision élevée), ainsi qu'une majorité de résultats positifs (rappel élevé).

## 7. Calibration

Lors de la classification, nous voulons souvent non seulement prédire l'étiquette de la classe, mais aussi obtenir une probabilité pour chaque étiquette. Cette probabilité devrait idéalement nous donner une certaine confiance dans la prédiction. Afin d'observer comment les probabilités générées par notre modèle sont alignées sur les probabilités réelles, nous pouvons tracer ce que l'on appelle une *courbe d'étalonnage*. 

Pour générer une courbe d'étalonnage, nous commençons par répartir nos prédictions dans un nombre fixe de cases (par exemple 5) entre 0 et 1. Nous calculons ensuite un point pour chaque case : la valeur x de chaque point est la moyenne de la probabilité que notre modèle a attribuée à ces points et la valeur y de chaque point est la fraction de vrais positifs dans cette case. Nous traçons ensuite ces points sur un graphique linéaire. Un modèle bien calibré a une courbe de calibrage qui s'aligne presque sur la ligne y=x.

La bibliothèque `sklearn` possède un utilitaire `calibration_curve` pour générer une courbe de calibration. Utilisons-le et regardons la calibration de notre modèle :

````python
from sklearn.calibration import calibration_curve
def plot_calibration_curve(y, pred):
    plt.figure(figsize=(20, 20))
    for i in range(len(class_labels)):
        plt.subplot(4, 4, i + 1)
        fraction_of_positives, mean_predicted_value = calibration_curve(y[:,i], pred[:,i], n_bins=20)
        plt.plot([0, 1], [0, 1], linestyle='--')
        plt.plot(mean_predicted_value, fraction_of_positives, marker='.')
        plt.xlabel("Predicted Value")
        plt.ylabel("Fraction of Positives")
        plt.title(class_labels[i])
    plt.tight_layout()
    plt.show()
````

````python
plot_calibration_curve(y, pred)
````

Comme le montrent les graphiques ci-dessus, pour la plupart des prédictions, le graphique de calibration de notre modèle ne ressemble pas à un graphique bien calibré. Comment y remédier ?...

Heureusement, il existe une méthode très utile appelée [Platt scaling](https://en.wikipedia.org/wiki/Platt_scaling) qui consiste à ajuster un modèle de régression logistique aux scores de notre modèle. Pour construire ce modèle, nous utiliserons la partie entraînement de notre ensemble de données pour générer le modèle linéaire, puis nous utiliserons le modèle pour calibrer les prédictions de notre partie test.

````python
from sklearn.linear_model import LogisticRegression as LR 

y_train = train_results[class_labels].values
pred_train = train_results[pred_labels].values
pred_calibrated = np.zeros_like(pred)

for i in range(len(class_labels)):
    lr = LR(solver='liblinear', max_iter=10000)
    lr.fit(pred_train[:, i].reshape(-1, 1), y_train[:, i])    
    pred_calibrated[:, i] = lr.predict_proba(pred[:, i].reshape(-1, 1))[:,1]
````

````python
plot_calibration_curve(y[:,], pred_calibrated)
````
